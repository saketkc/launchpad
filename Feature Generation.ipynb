```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("train.csv") 
```

```python
data
```

```python
# Extract Id(Hour Id), Ref, Kdp, Expected 
important_columns = data[['Id','Ref','Kdp','Expected']]
```

```python
store = pd.HDFStore('store.h5')
```

```python
important_columns.to_hdf('store.h5','table',append=True)
```

```python
important_columns = pd.read_hdf('store.h5','table',where = ['index<50000']) 
```

```python
# Get cuttoff points for 10 bins for reflectivity column
reflectivity_cutoff = important_columns['Ref'].quantile([.1, .2, .3, .4, .5, .6, .7, .8, .9])
```

```python
# Get cuttoff points for 10 bins for Kdp column
kdp_cutoff = important_columns['Kdp'].quantile([.1, .2, .3, .4, .5, .6, .7, .8, .9])
```

```python
groupedById = important_columns.groupby(['Id'])
```

```python
columns_set1 = ['totalHourlyScans','Ylabel']
features_set1 = pd.DataFrame(columns = columns_set1)
```

```python
for i, group in groupedById:
    numberOfScans = len(group['Expected'])
    ylabel = group['Expected'].iloc[0]
    features_set1.loc[i] = [numberOfScans,ylabel]
```

```python
features_set1
```

```python
columns_set2 = ['ref_nan','ref_1','ref_2','ref_3','ref_4','ref_5','ref_6','ref_7','ref_8','ref_9','ref_10']
features_set2 = pd.DataFrame(columns = columns_set2)
```

```python
import math
for i, group in groupedById:
    ref_array = [0] * 11
    refGroup = group['Ref']
    for val in refGroup:
        if(math.isnan(val)):
            ref_array[0] = 1
        if(val <= reflectivity_cutoff.iloc[0]):
            ref_array[1] = 1
        if(val > reflectivity_cutoff.iloc[0] and val <= reflectivity_cutoff.iloc[1]):
            ref_array[2] = 1
        if(val > reflectivity_cutoff.iloc[1] and val <= reflectivity_cutoff.iloc[2]):
            ref_array[3] = 1
        if(val > reflectivity_cutoff.iloc[2] and val <= reflectivity_cutoff.iloc[3]):
            ref_array[4] = 1
        if(val > reflectivity_cutoff.iloc[3] and val <= reflectivity_cutoff.iloc[4]):
            ref_array[5] = 1
        if(val > reflectivity_cutoff.iloc[4] and val <= reflectivity_cutoff.iloc[5]):
            ref_array[6] = 1
        if(val > reflectivity_cutoff.iloc[5] and val <= reflectivity_cutoff.iloc[6]):
            ref_array[7] = 1
        if(val > reflectivity_cutoff.iloc[6] and val <= reflectivity_cutoff.iloc[7]):
            ref_array[8] = 1
        if(val > reflectivity_cutoff.iloc[7] and val <= reflectivity_cutoff.iloc[8]):
            ref_array[9] = 1
        if(val > reflectivity_cutoff.iloc[8]):
            ref_array[10] = 1
    features_set2.loc[i] = ref_array
```

```python
features_set2
```

```python
columns_set3 = ['kdp_nan','kdp_1','kdp_2','kdp_3','kdp_4','kdp_5','kdp_6','kdp_7','kdp_8','kdp_9','kdp_10']
features_set3 = pd.DataFrame(columns = columns_set3)
```

```python
import math
for i, group in groupedById:
    kdp_array = [0] * 11
    kdpGroup = group['Kdp']
    for val in kdpGroup:
        if(math.isnan(val)):
            kdp_array[0] = 1
        if(val <= kdp_cutoff.iloc[0]):
            kdp_array[1] = 1
        if(val > kdp_cutoff.iloc[0] and val <= kdp_cutoff.iloc[1]):
            kdp_array[2] = 1
        if(val > kdp_cutoff.iloc[1] and val <= kdp_cutoff.iloc[2]):
            kdp_array[3] = 1
        if(val > kdp_cutoff.iloc[2] and val <= kdp_cutoff.iloc[3]):
            kdp_array[4] = 1
        if(val > kdp_cutoff.iloc[3] and val <= kdp_cutoff.iloc[4]):
            kdp_array[5] = 1
        if(val > kdp_cutoff.iloc[4] and val <= kdp_cutoff.iloc[5]):
            kdp_array[6] = 1
        if(val > kdp_cutoff.iloc[5] and val <= kdp_cutoff.iloc[6]):
            kdp_array[7] = 1
        if(val > kdp_cutoff.iloc[6] and val <= kdp_cutoff.iloc[7]):
            kdp_array[8] = 1
        if(val > kdp_cutoff.iloc[7] and val <= kdp_cutoff.iloc[8]):
            kdp_array[9] = 1
        if(val > kdp_cutoff.iloc[8]):
            kdp_array[10] = 1
    features_set3.loc[i] = kdp_array
```

```python
features_set3
```

```python
frames = [features_set2, features_set3, features_set1]
```

```python
combined_features = pd.concat(frames,axis = 1)
```

```python
combined_features
```

```python
combined_features.to_csv('features_generated.csv')
```
